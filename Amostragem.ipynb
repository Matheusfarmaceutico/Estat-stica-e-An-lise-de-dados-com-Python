```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, amos_estra = train_test_split(enem_campinas.drop(columns = ["RACA"]),
                                               enem_campinas["RACA"],
                                               stratify=enem_campinas["RACA"],
                                                test_size=0.20)
amos_estra.value_counts()
```



SOBRE O QUE ESTÁ ACIMA: Essa função train_test_split obrigatoriamente necessita de três parâmetros principais que são chamados de X, Y e test_size. Como só nos interessa o parâmetro Y_teste (amos_estra), eu coloquei para o parâmetro X dados quaisquer. Eu escolhi colocar todas as variáveis, exceto a coluna Raça, por isso foi feito o drop. Poderia ser quaisquer outras colunas, esse parâmetro não nos serve pra nada, eu só "enganei" a função colocando outras colunas, o que nos interessa é o parâmetro Y

Parâmetro X: enem_campinas.drop('RACA', axis=1)

Para o parâmetro Y foi utilizado somente a coluna Raça. Essa coluna é a que eu vou fazer a amostragem.

Parâmetro Y: enem_campinas['RACA']

O parâmetro stratify é para indicar que será realizada uma estratificação exatamente na coluna RACA (que nada mais é do que o parâmetro Y).

stratify=enem_campinas['RACA'] E por fim, o parâmetro test_size é para indicar a porcentagem que será estratificada.

Resumindo, foi utilizada uma função que não foi projetada para fazer amostragem e por este motivo tivemos que fazer umas adaptações.

O parâmetro X, que recebeu o drop, na verdade não foi utilizado, mas como é obrigatório colocá-lo eu coloquei sem a variável RACA, mas você pode colocar outra coisa, inclusive sem dar um drop na coluna RACA, o resultado será o mesmo.
