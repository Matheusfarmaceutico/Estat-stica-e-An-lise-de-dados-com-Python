# **Amostragem Aleatória Simples**

```python
# Criando semente aleatória, para fixar o resultado.
np.random.seed(1)
# Cada número que eu seto firmará um resultado.
#np.random.seed(2) me daria um outro resultado, mas bastava inserir np.random.seed(1) para visualizar o resultado anterior.
```

```python
amos_simples = enem_campinas.sample(frac = 0.20) #Porcentagem
```

```python
amos_simples
```

```python
amos_simples = enem_campinas.sample(3000) #número inteiro
```
