```python
import pandas as pd
import numpy as np
from collections import Counter
pd.options.display.max_columns=200
```

```python
enem_sp = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/05 - Distribuição de Frequência/02 - Ex-distribuicao_frequencia/enem_tratado.csv", sep=",", encoding="iso-8859-1")
```

```python
enem_sp.head()
```

```python
enem_sorocaba = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == "Sorocaba"]
```

```python
enem_sorocaba.head()
```

```python
enem_sorocaba.shape
```

```python
idade = enem_sorocaba["IDADE"]
```

```python
idade
```

```python
freq_absoluta = Counter(idade)
```

```python
freq_absoluta
```

```python
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient="index")
```

```python
freq_absoluta
```

```python
freq_absoluta = freq_absoluta.sort_index(ascending=True)
freq_absoluta
```

```python
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa
```

```python
#tirando a prova = soma = 1.0
freq_relativa.sum()
```

```python
freq_perc_relativa = freq_relativa * 100
freq_perc_relativa
```

#Histograma

```python
import matplotlib.pyplot as plt
```

```python
enem_sorocaba.IDADE.hist(bins=9) #seta o intervalo de idades
plt.title("Distribuição de Idades - Enem SP (2019) - Sorocaba")
plt.show()
```

```python
#Outra forma de fazer o mesmo

plt.hist(enem_sorocaba.IDADE,8,rwidth=1) # 9 é o bins; rwidth é para aumentar o espaçamento entre as barras (0-1)
plt.title("Distribuição de Idades - Enem SP (2019) - Sorocaba")
plt.show()
```

```python
from matplotlib import colors
```

```python
plt.hist(enem_sorocaba.IDADE,10,rwidth=1,color="red",density=True) #Density faz com que deixe de ser contagem
plt.title("Distribuição de Idades - Enem SP (2019) - Sorocaba")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()


# Faixa de idade com bins=10

# 17-20 (diferenca = 3)
# 3x0.14 = 0,14

# A soma de cada um deles tem que dar 1.
```
